```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import time

def get_coordinates(country):
    geolocator = Nominatim(user_agent="my_geocoding_app")  
    try:
        location = geolocator.geocode(country)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        # Wait for a short period before retrying
        time.sleep(1)
        return get_coordinates(country)
    except Exception as e:
        print(f"Error: {e}")
        return None, None
```

```python
import seaborn as sns

# Melt the DataFrame to have 'Salary Type' and 'Value' columns
df_melted = salary_dataset.melt(id_vars=['Country'], value_vars=['Median Salary', 'Average Salary', 'Lowest Salary', 'Highest Salary'],
                     var_name='Salary Type', value_name='Value')

# Define a function to remove outliers
def remove_outliers(df):
    Q1 = df['Value'].quantile(0.25)
    Q3 = df['Value'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df['Value'] >= lower_bound) & (df['Value'] <= upper_bound)]

# Remove outliers from the melted DataFrame
df_no_outliers = remove_outliers(df_melted)

# Plotting the box plot without outliers
plt.figure(figsize=(12, 8))
sns.boxplot(x='Salary Type', y='Value', data=df_no_outliers)
plt.title('Distribution of Salaries by Type (Without Outliers)')
plt.xlabel('Salary Type')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
country = input("Enter the country name for GDP per capita forecast: ")
df_country = filtered_data[filtered_data['Country'] == country]

if df_country.empty:
    print(f"No data available for {country}.")
else:
    df_country = df_country[['Year', 'GDP per capita']]
    
    # Filter out rows where 'Value' is 0
    df_country = df_country[df_country['GDP per capita'] != 0]
    
    if df_country.empty:
        print(f"No valid data available for {country} after filtering out zero values.")
    else:
        df_country.set_index('Year', inplace=True)
        
        try:
            modelAR = sm.tsa.ARIMA(df_country, order=(1, 1, 1))
            resultsAR = modelAR.fit()
            forecast = resultsAR.get_forecast(steps=10).predicted_mean
            
            future_years = pd.DataFrame({
                'Year': range(df_country.index[-1] + 1, df_country.index[-1] + 11),
                'GDP per capita': forecast
            }).set_index('Year')
            
            plt.figure(figsize=(10, 6))
            plt.plot(df_country, label='Historical')
            plt.plot(future_years, label='Forecast', linestyle='--')
            plt.title(f'GDP per capita Forecast for {country}')
            plt.xlabel('Year')
            plt.ylabel('GDP per capita')
            plt.legend()
            plt.xticks(rotation=45)
            plt.gca().xaxis.set_major_locator(plt.MaxNLocator(integer=True))
            plt.grid(True)
            plt.tight_layout()
            plt.show()
            
            print(future_years)
        except Exception as e:
            print(f"An error occurred while fitting the model or forecasting: {e}")
```

```python
country = input("Enter the country name for GDP per capita forecast: ")
df_country = filtered_data[filtered_data['Country'] == country]
if df_country.empty:
    print(f"No data available for {country}.")
else:
    df_country = df_country[['Year', 'GDP per capita']]
    df_country.set_index('Year', inplace=True)
    df_country = df_country[df_country['GDP per capita'] != 0]

    if df_country.empty:
        print(f"No valid data available for {country} after filtering out zero values.")
    else:
        # Hyperparameter tuning
        p = d = q = range(0, 3)
        pdq = list(itertools.product(p, d, q))
        aic_values = []

        for param in pdq:
            try:
                model = sm.tsa.ARIMA(df_country, order=param)
                results = model.fit()
                aic_values.append((param, results.aic))
            except:
                continue

        best_params = min(aic_values, key=lambda x: x[1])[0]
        print(f"Best ARIMA parameters: {best_params}")

        # Fit the best model
        modelBestAR = sm.tsa.ARIMA(df_country, order=best_params)
        resultsBestAR = modelBestAR.fit()

        # Forecast
        forecast = results.get_forecast(steps=10)
        forecast_mean = forecast.predicted_mean
        forecast_ci = forecast.conf_int()

        # Create a DataFrame for future dates
        future_years = pd.DataFrame({
            'Year': range(df_country.index[-1] + 1, df_country.index[-1] + 11),
            'GDP per capita': forecast_mean
        }).set_index('Year')

        # Plot the forecast with confidence intervals
        plt.figure(figsize=(14, 8))
        plt.plot(df_country.index, df_country['GDP per capita'], label='Historical Data')
        plt.plot(future_years.index, future_years['GDP per capita'], label='Forecast', linestyle='--')
        plt.fill_between(future_years.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='gray', alpha=0.2)
        plt.title(f'GDP per capita Forecast for {country}', fontsize=16)
        plt.xlabel('Year', fontsize=14)
        plt.ylabel('GDP per capita', fontsize=14)
        plt.xticks(np.arange(df_country.index.min(), future_years.index.max() + 1, 5))  # Display year every 5 years
        plt.legend(loc='upper left', fontsize=12)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        print(future_years)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

```python
# Prompt the user to input a country name
country = input("Enter the country name for GDP per capita forecast: ")

# Filter data for the specified country
df_country2 = income_df[income_df['Country'] == country]

# Check if the country exists in the dataset
if df_country2.empty:
    print(f"No data available for {country}.")
else:
    # Prepare the data
    df_country2 = df_country2[['Year', 'Value']]
    df_country2.set_index('Year', inplace=True)

    # Ensure data is sorted by Year
    df_country2 = df_country2.sort_index()

    # Hyperparameter tuning with cross-validation
    p = d = q = range(0, 3)
    pdq = list(itertools.product(p, d, q))
    aic_values = []

    for param in pdq:
        try:
            model2BestAR = sm.tsa.ARIMA(df_country2, order=param)
            results2BestAR = model2BestAR.fit()
            aic_values.append((param, results2BestAR.aic))
        except:
            continue

    best_params = min(aic_values, key=lambda x: x[1])[0]
    print(f"Best ARIMA parameters: {best_params}")

    # Fit the best model
    model2BestAR = sm.tsa.ARIMA(df_country2, order=best_params)
    results2BestAR = model2BestAR.fit()

    # Forecast
    forecast = results2BestAR.get_forecast(steps=10)
    forecast_mean = forecast.predicted_mean
    forecast_ci = forecast.conf_int()

    # Create a DataFrame for future dates
    future_years = pd.DataFrame({
        'Year': range(df_country2.index[-1] + 1, df_country2.index[-1] + 11),
        'Value': forecast_mean
    }).set_index('Year')

    # Plot the forecast with confidence intervals
    plt.figure(figsize=(14, 8))
    plt.plot(df_country2.index, df_country2['Value'], label='Historical Data', color='blue')
    plt.plot(future_years.index, future_years['Value'], label='Forecast', linestyle='--', color='orange')
    plt.fill_between(future_years.index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='gray', alpha=0.2)
    plt.title(f'Inequality in Income Forecast for {country}', fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Value', fontsize=14)
    plt.xticks(df_country2.index.union(future_years.index))  # Ensure all years are included in x-axis
    plt.legend(loc='upper left', fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Display the forecasted values
    print(future_years)
```
